```python
#Q1. Creating a DataFrame
#Create a Pandas DataFrame using a Python dictionary with the following columns:
#Product
#Region
#Sales
#Quantity
#Add at least 5 rows of sample data and display the DataFrame.
#Dataset for Questions 2–10
#Use the dataset file provided:
#answer -
import pandas as pd

# Creating a sample dictionary
data = {
    "Product": ["Laptop", "Tablet", "Phone", "Monitor", "Keyboard"],
    "Region": ["West", "South", "East", "North", "West"],
    "Sales": [1056, 904, 2028, 1269, 777],
    "Quantity": [2, 7, 4, 8, 8]
}

# Creating DataFrame
df_sample = pd.DataFrame(data)
df_sample
```

```python
#Q2. Data Inspection
#Perform the following inspection tasks:
#1 Display the first 5 rows of the dataset.
#2 Display the last 5 rows.
#3 Show the structure of the dataset using info().
#4 Display the shape of the dataset
#Answer- Load dataset
df = pd.read_csv("/content/Retail_dataset.csv")

# 1. First 5 rows
df.head()

# 2. Last 5 rows
df.tail()

# 3. Structure of dataset
df.info()

# 4. Shape of dataset
df.shape

#Output (summarized):
#- First 5 rows → shows Sneha, Neha, Meera, Rohan, Priya entries
#- Last 5 rows → shows rows up to Order_ID 1090
#- info() → 6 columns, 90 entries, all non-null
#- Shape → (90, 6)
```

```python
#Q3. Column Selection
#Select and display only the following columns:
#Product
#Region
#Sales
#Answer -
df[["Product", "Region", "Sales"]]
#Output: Displays only those three columns.
```

```python
#Q4. Indexing using .loc
#Using .loc, perform the following:
#Display the row with index 10.
#Display the Sales value of the row with index 20
#Answer-# 1. Row with index 10
df.loc[10]
# 2. Sales value of row with index 20
df.loc[20, "Sales"]
#Output:
#- Row 10 → Order_ID=1011, Customer=Raj, Product=Monitor, Region=North, Sales=1693, Quantity=6
#- Sales at index 20 → 1503
```

```python
#Q5. Indexing using .iloc
#Using .iloc, display:
#1 The first 6 rows and first 4 columns.
#2 The Sales value from the 3rd row.
#Answer - # 1. The first 6 rows and first 4 columns
print("First 6 rows and 4 columns:")
print(df.iloc[:6, :4])
# 2. The Sales value from the 3rd row (index 2)
# Column index for 'Sales' is 4 (Order_ID=0, Customer=1, Product=2, Region=3, Sales=4)
sales_3rd_row = df.iloc[2, 4]
print(f"\nSales value from the 3rd row: {sales_3rd_row}")
#Output:
#- First 6 rows with first 4 columns → shows Order_ID, Customer, Product, Region
#- Sales from 3rd row → 1597
```

```python
#Q6. Row Slicing
#Display:
#1The first 10 rows of the dataset.
#2Rows 15 to 25.
#Answer # 1. The first 10 rows of the dataset
print("First 10 rows:")
print(df[0:10])
# 2. Rows 15 to 25
print("\nRows 15 to 25:")
print(df[15:26])
```

```python
#Q7. Filtering Data
#Filter the dataset to display:
#1Rows where Region = "East".
#2Rows where Sales > 1500.
#Answer-# 1. Rows where Region = "East"
east_region = df[df['Region'] == 'East']
print("Rows in East region:")
print(east_region.head())
# 2. Rows where Sales > 1500
high_sales = df[df['Sales'] > 1500]
print("\nRows with Sales > 1500:")
print(high_sales.head())
```

```python
#Q8. Multiple Condition Filtering
#Display rows where:
#Region is West
#Sales is greater than 1200
#Answer-# Filtering with multiple conditions
west_high_sales = df[(df['Region'] == 'West') & (df['Sales'] > 1200)]
west_high_sales
```

```python
#Q9. Descriptive Statistics
#Generate summary statistics using:
#describe()
#From the result identify:
#Mean sales value
#Maximum quantity sold
#Minimum sales value
#Answer- # Generate summary statistics
stats = df.describe()
print("Summary Statistics:")
print(stats)
# Identifying specific values from the description
mean_sales = df['Sales'].mean()
max_qty = df['Quantity'].max()
min_sales = df['Sales'].min()
print(f"\nMean Sales Value: {mean_sales}")
print(f"Maximum Quantity Sold: {max_qty}")
print(f"Minimum Sales Value: {min_sales}")
#Output:
#- Mean Sales ≈ 1406
#- Max Quantity = 8
#- Min Sales = 515
```

```python
#Q10. Aggregation Analysis
#Using the Sales column, calculate:
#Total sales
#Average sales
#Maximum sales
#Minimum sales
#Answer-# Aggregation calculations on the Sales column
total_sales = df['Sales'].sum()
avg_sales = df['Sales'].mean()
max_sales = df['Sales'].max()
min_sales = df['Sales'].min()
print(f"Total Sales: {total_sales}")
print(f"Average Sales: {avg_sales}")
print(f"Maximum Sales: {max_sales}")
print(f"Minimum Sales: {min_sales}")
#Output:
#- Total Sales = 126620
#- Average Sales ≈ 1406
#- Max Sales = 2455
#- Min Sales = 515
```
