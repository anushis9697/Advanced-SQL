```python
# Importing required libraries
import pandas as pd
import numpy as np
```

```python
#Q1.Load the given retail sales dataset into a Pandas DataFrame and display the first 5 rows along with the dataset information to understand its structure
#Answer-
# Load the dataset
df = pd.read_csv('retail_dataset .csv')
# Display the first 5 rows
display(df.head())
# Display dataset information to understand structure
df.info()
#Output Value:
#Head: Displays transactions T001 to T005.
#Info: Shows 100 entries with columns: Transaction ID, Customer ID, Product Category, Quantity, Price, Date, and Region.
```

```python
#Q2. Inspect the dataset and identify missing values in each column. Display the total number of missing values present in the dataset
#Answer- # Missing values per column
missing_per_column = df.isnull().sum()
print("Missing values per column:")
print(missing_per_column)
# Total missing values
total_missing = df.isnull().sum().sum()
print(f"\nTotal missing values in dataset: {total_missing}")
#Output:
#Transaction ID       0
#Customer ID          0
#Product Category     3
#Quantity             2
#Price                6
#Date                 2
#Region               2
#Total missing values: 15
```

```python
#Q3. Replace missing values in the Quantity and Price columns with their respective mean values using Pandas.
#Answer-# Filling missing Quantity and Price with mean
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())
df['Price'] = df['Price'].fillna(df['Price'].mean())

print("Missing values after imputation:")
print(df[['Quantity', 'Price']].isnull().sum())
#Output Value: Quantity: 0, Price: 0.
```

```python
#Q4. Remove all rows where the Product Category or Region column contains missing values.
#Answer-
df_clean = df.dropna(subset=['Product Category','Region'])

print("Rows after cleaning:", df_clean.shape[0])
#Output Value: Remaining rows: 95
```

```python
#Q5. Create a new column called Revenue using NumPy, where:
#Revenue = Quantity × Price
#Display the updated dataset with the new column.
#Answer-
df_clean['Revenue'] = df_clean['Quantity'] * df_clean['Price']
display(df_clean.head())
#Output (first 5 rows with Revenue)
```

```python
#Q6. Using NumPy, calculate the total revenue generated from all transactions in the dataset.
#Answer-
total_revenue = df_clean['Revenue'].sum()
print("Total Revenue:", total_revenue)
```

```python
#Q7. Group the dataset by Product Category and calculate the total revenue generated by each category.
#Answer-
category_revenue = df_clean.groupby('Product Category')['Revenue'].sum().sort_values(ascending=False)
print(category_revenue)
```

```python
#Q8. Based on the revenue calculated for each product category, identify the top 3 categories with the highest revenue and the bottom 3 categories with the lowest revenue.
#Answer-
top3 = category_revenue.head(3)
bottom3 = category_revenue.tail(3)
print("Top 3 Categories:\n", top3)
print("\nBottom 3 Categories:\n", bottom3)
```

```python
#Q9. Group the dataset by Region and calculate the total revenue generated in each region. Identify the region with the highest revenue and the region with the lowest revenue
#Answer-
region_revenue = df_clean.groupby('Region')['Revenue'].sum().sort_values(ascending=False)
print(region_revenue)
print("\nHighest Revenue Region:", region_revenue.idxmax())
print("Lowest Revenue Region:", region_revenue.idxmin())
```

```python
#Q10. Extract the month from the Date column using Pandas and group the dataset by month to calculate the
#total monthly revenue. Also, calculate the mean, median, and standard deviation of the Revenue column
#using NumPy
#Answer
df_clean['Date'] = pd.to_datetime(df_clean['Date'], dayfirst=True, errors='coerce')
df_clean['Month'] = df_clean['Date'].dt.month

monthly_revenue = df_clean.groupby('Month')['Revenue'].sum()
print("Monthly Revenue:\n", monthly_revenue)

mean_rev = np.mean(df_clean['Revenue'])
median_rev = np.median(df_clean['Revenue'])
std_rev = np.std(df_clean['Revenue'])

print("\nRevenue Statistics:")
print("Mean:", mean_rev)
print("Median:", median_rev)
print("Standard Deviation:", std_rev)
```
